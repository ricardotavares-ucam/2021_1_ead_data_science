### Importando a biblioteca Pandas

```python
import pandas as pd

# Formatação de pontos flutuantes
pd.options.display.float_format = '{:,.4f}'.format
```

### Criando os dados através de um dicionário de dados

```python
dados = {'Feminino': 58, 'Masculino': 40, 'Não declarado': 2}
```

### Transformando o dicionário em um DataFrame

```python
df = pd.DataFrame.from_dict(dados, orient='index', columns=['Freq. Absoluta'])
```

```python
df
```

### Calculando o somatório das quantidades

```python
total = df['Freq. Absoluta'].sum()
total
```

# Frequência absoluta e Frequência Relativa

### Calculando a Frequência relativa de cada variável nominal

```python
df['Freq. Relativa'] = (df['Freq. Absoluta']/total)
```

```python
df['Freq. Relativa %'] = (df['Freq. Absoluta']/total)*100
```

```python
df
```

### Importando a biblioteca Matplotlib para plotagem de gráficos

### Distribuição de frequências relativas e acumuladas

```python
dados = {'Fevereiro': 19, 'Março': 6, 'Abril': 6, 'Maio': 11, 'Junho': 23, 'Julho': 20, 'Agosto': 18}
```

```python
df = pd.DataFrame.from_dict(dados, orient='index', columns=['Freq. Absoluta'])
```

```python
df
```

```python
total = df['Freq. Absoluta'].sum()
total
```

```python
df['Freq. Relativa'] = (df['Freq. Absoluta']/total)
```

```python
df['Freq. Relativa %'] = (df['Freq. Absoluta']/total)*100
```

```python
df
```

```python
df['Freq. Absoluta Acum.'] = df['Freq. Absoluta'].cumsum()
```

```python
df
```

```python
df['Freq. Relativa Acum. %'] = df['Freq. Relativa %'].cumsum()
```

```python
df
```

```python
plt.bar(df.index, df['Freq. Absoluta'], label = 'Número de Entrevistados')
plt.title('Gráfico de Entrevistas por meses', fontsize=16, fontweight='bold')
plt.legend()
plt.ylabel('Nº Entrevistados', fontsize=14)
plt.xlabel('Meses', fontsize=14)
for i, v in enumerate(df['Freq. Absoluta']):
    plt.text(i, v + .25, str(v), fontsize=14)
plt.show()
```

# Medidas de Posição

### Média
### Mediana
### Moda


```python
# Dado o conjunto de dados (idade), calcule a média, mediana e moda
# 22 idades
x = [4, 10, 20, 21, 26, 35, 48, 45, 49, 15, 11, 18, 17, 5, 2, 14, 88, 79, 55, 64, 47, 49]

df = pd.DataFrame(x, columns =['Idades']) 
df
```

```python
print("\n----------- Calculando a Média -----------\n")
print(df['Idades'].mean())
 
print("\n----------- Calculando a Mediana -----------\n")
print(df.Idades.median())
 
print("\n----------- Calculando a Moda -----------\n")
print(df.mode())
```

```python
df.value_counts()
```

# Medidas de Dispersão

### Amplitude

```python
amplitude = df['Idades'].max() - df['Idades'].min()
print("\n----------- Calculando a Amplitudade -----------\n")
amplitude
```

```python
df.sort_values(by=['Idades'])
```

### Variância

```python
print("\n----------- Calculando a Variância -----------\n")
df['Idades'].var()
```

### Desvio Padrão

```python
print("\n----------- Calculando o Desvio Padrão -----------\n")
df['Idades'].std()
```

```python
print("\n----------- Calculando o Coef. de Variação -----------\n")
cv = df['Idades'].std()/df['Idades'].mean()
cv
```
